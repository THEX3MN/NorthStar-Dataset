```r
# Joining the tables
# Orders joined with Customers

orders_customers <- sqldf("
SELECT
    o.order_id,
    o.customer_id,
    o.service_type,
    o.order_created_at,
    o.promised_window_hours,
    o.pickup_zone,
    o.dropoff_zone,
    o.priority_level,
    o.order_value,
    o.booking_channel,
    o.special_handling_flag,
    c.age,
    c.home_zone,
    c.customer_type,
    c.loyalty_score,
    c.app_engagement_score,
    c.preferred_channel,
    c.account_status
FROM orders o
LEFT JOIN customers c
ON o.customer_id = c.customer_id
")

# Incidents joined with Deliveries

incidents_deliveries <- sqldf("
SELECT
    i.incident_id,
    i.delivery_id,
    i.incident_type,
    i.reported_at,
    i.severity,
    i.resolution_status,
    i.resolved_hours,
    d.order_id,
    d.driver_id,
    d.vehicle_id,
    d.hub_id,
    d.dispatch_time,
    d.delivery_completed_at,
    d.delivery_status,
    d.route_distance_km,
    d.manual_route_override_count,
    d.proof_of_completion_missing,
    d.customer_rating_post_delivery,
    d.fuel_or_charge_cost
FROM incidents i
LEFT JOIN deliveries d
ON i.delivery_id = d.delivery_id
")

# Checking datasets after joining

head(orders_customers)
head(incidents_deliveries)
```

```r
# Building queries
# Orders and Customers view

orders_customer_view <- sqldf("
SELECT
    c.customer_type,
    COUNT(o.order_id) AS total_orders,
    ROUND(AVG(o.order_value), 2) AS avg_order_value,
    ROUND(AVG(o.promised_window_hours), 2) AS avg_promised_window,
    COUNT(CASE WHEN o.booking_channel = 'App' THEN 1 END) AS app_bookings,
    COUNT(CASE WHEN o.booking_channel = 'Phone' THEN 1 END) AS phone_bookings,
    COUNT(CASE WHEN o.booking_channel = 'Web' THEN 1 END) AS web_bookings
FROM orders o
LEFT JOIN customers c
ON o.customer_id = c.customer_id
GROUP BY c.customer_type
ORDER BY total_orders DESC
")
orders_customer_view

# Delivery Performance view

delivery_performance_view <- sqldf("
SELECT
    delivery_status,
    COUNT(delivery_id) AS total_deliveries,
    ROUND(AVG(route_distance_km), 2) AS avg_route_distance,
    ROUND(AVG(manual_route_override_count), 2) AS avg_manual_overrides,
    ROUND(AVG(proof_of_completion_missing), 2) AS avg_missing_proof,
    ROUND(AVG(customer_rating_post_delivery), 2) AS avg_customer_rating,
    ROUND(AVG(fuel_or_charge_cost), 2) AS avg_fuel_or_charge_cost
FROM deliveries
GROUP BY delivery_status
ORDER BY total_deliveries DESC
")
delivery_performance_view

# Incident view

incident_view <- sqldf("
SELECT
    incident_type,
    severity,
    resolution_status,
    COUNT(incident_id) AS total_incidents,
    ROUND(AVG(resolved_hours), 2) AS avg_resolved_hours
FROM incidents
GROUP BY incident_type, severity, resolution_status
ORDER BY total_incidents DESC
")
incident_view

# Complaint view

complaint_view <- sqldf("
SELECT
    complaint_type,
    severity,
    status,
    COUNT(complaint_id) AS total_complaints,
    ROUND(AVG(resolution_days), 2) AS avg_resolution_days,
    ROUND(AVG(compensation_amount), 2) AS avg_compensation_amount
FROM complaints
GROUP BY complaint_type, severity, status
ORDER BY total_complaints DESC
")
complaint_view

# Zone Comparison view

zone_comparison_view <- sqldf("
SELECT
    h.zone,
    COUNT(d.delivery_id) AS total_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Ontime' THEN 1 ELSE 0 END) AS ontime_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Delayed' THEN 1 ELSE 0 END) AS delayed_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Failed' THEN 1 ELSE 0 END) AS failed_deliveries,
    ROUND(AVG(d.route_distance_km), 2) AS avg_route_distance,
    ROUND(AVG(d.customer_rating_post_delivery), 2) AS avg_customer_rating,
    ROUND(AVG(d.fuel_or_charge_cost), 2) AS avg_fuel_or_charge_cost
FROM deliveries d
LEFT JOIN hubs h
ON d.hub_id = h.hub_id
GROUP BY h.zone
ORDER BY total_deliveries DESC
")
zone_comparison_view

# Hubs Comparison View

hub_comparison_view <- sqldf("
SELECT
    h.hub_id,
    h.hub_name,
    h.zone,
    h.hub_type,
    h.capacity_score,
    COUNT(d.delivery_id) AS total_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Ontime' THEN 1 ELSE 0 END) AS ontime_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Delayed' THEN 1 ELSE 0 END) AS delayed_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Failed' THEN 1 ELSE 0 END) AS failed_deliveries,
    ROUND(AVG(d.customer_rating_post_delivery), 2) AS avg_customer_rating,
    ROUND(AVG(d.manual_route_override_count), 2) AS avg_manual_overrides
FROM deliveries d
LEFT JOIN hubs h
ON d.hub_id = h.hub_id
GROUP BY h.hub_id, h.hub_name, h.zone, h.hub_type, h.capacity_score
ORDER BY total_deliveries DESC
")
hub_comparison_view

# Query optimisation

# Complaints by order
complaints_by_order <- sqldf("
SELECT
    order_id,
    COUNT(complaint_id) AS complaint_count,
    SUM(resolution_days) AS total_resolution_days,
    AVG(resolution_days) AS avg_resolution_days
FROM complaints
GROUP BY order_id
")

# Deliveries by order
deliveries_by_order <- sqldf("
SELECT
    order_id,
    COUNT(delivery_id) AS delivery_count,
    SUM(customer_rating_post_delivery) AS total_customer_rating,
    AVG(customer_rating_post_delivery) AS avg_customer_rating
FROM deliveries
GROUP BY order_id
")

# Orders by customer
orders_by_customer <- sqldf("
SELECT
    customer_id,
    COUNT(order_id) AS order_count,
    SUM(order_value) AS total_order_value,
    AVG(order_value) AS avg_order_value
FROM orders
GROUP BY customer_id
")

# Complaints by customer
complaints_by_customer <- sqldf("
SELECT
    customer_id,
    COUNT(complaint_id) AS complaint_count,
    SUM(resolution_days) AS total_resolution_days,
    AVG(resolution_days) AS avg_resolution_days
FROM complaints
GROUP BY customer_id
")

# Deliveries by customer
deliveries_by_customer <- sqldf("
SELECT
    o.customer_id,
    COUNT(d.delivery_id) AS delivery_count,
    SUM(d.customer_rating_post_delivery) AS total_customer_rating,
    AVG(d.customer_rating_post_delivery) AS avg_customer_rating
FROM orders o
LEFT JOIN deliveries d
ON o.order_id = d.order_id
GROUP BY o.customer_id
")

# Service type comparison

service_type_comparison_view <- sqldf("
SELECT
    o.service_type,
    COUNT(DISTINCT o.order_id) AS total_orders,
    ROUND(AVG(o.order_value), 2) AS avg_order_value,
    ROUND(AVG(o.promised_window_hours), 2) AS avg_promised_window,
    COALESCE(SUM(cbo.complaint_count), 0) AS total_complaints,
    ROUND(
        CASE
            WHEN SUM(cbo.complaint_count) > 0
            THEN SUM(cbo.total_resolution_days) / SUM(cbo.complaint_count)
        END, 2
    ) AS avg_complaint_resolution_days,
    COALESCE(SUM(dbo.delivery_count), 0) AS total_deliveries,
    ROUND(
        CASE
            WHEN SUM(dbo.delivery_count) > 0
            THEN SUM(dbo.total_customer_rating) / SUM(dbo.delivery_count)
        END, 2
    ) AS avg_customer_rating
FROM orders o
LEFT JOIN complaints_by_order cbo
ON o.order_id = cbo.order_id
LEFT JOIN deliveries_by_order dbo
ON o.order_id = dbo.order_id
GROUP BY o.service_type
ORDER BY total_orders DESC
")
service_type_comparison_view

# Customer type comparison

customer_type_performance_view <- sqldf("
SELECT
    c.customer_type,
    COUNT(DISTINCT c.customer_id) AS total_customers,
    SUM(IFNULL(oc.order_count, 0)) AS total_orders,
    ROUND(
        CASE
            WHEN SUM(IFNULL(oc.order_count, 0)) > 0
            THEN SUM(IFNULL(oc.total_order_value, 0)) / SUM(IFNULL(oc.order_count, 0))
        END, 2
    ) AS avg_order_value,
    SUM(IFNULL(cc.complaint_count, 0)) AS total_complaints,
    ROUND(
        CASE
            WHEN SUM(IFNULL(cc.complaint_count, 0)) > 0
            THEN SUM(IFNULL(cc.total_resolution_days, 0)) / SUM(IFNULL(cc.complaint_count, 0))
        END, 2
    ) AS avg_complaint_resolution_days,
    SUM(IFNULL(dc.delivery_count, 0)) AS total_deliveries,
    ROUND(
        CASE
            WHEN SUM(IFNULL(dc.delivery_count, 0)) > 0
            THEN SUM(IFNULL(dc.total_customer_rating, 0)) / SUM(IFNULL(dc.delivery_count, 0))
        END, 2
    ) AS avg_customer_rating
FROM customers c
LEFT JOIN orders_by_customer oc
ON c.customer_id = oc.customer_id
LEFT JOIN complaints_by_customer cc
ON c.customer_id = cc.customer_id
LEFT JOIN deliveries_by_customer dc
ON c.customer_id = dc.customer_id
GROUP BY c.customer_type
ORDER BY total_orders DESC
")
customer_type_performance

# Driver performance view

driver_performance_view <- sqldf("
SELECT
    dr.driver_id,
    dr.base_zone,
    dr.employment_type,
    dr.years_experience,
    dr.training_score,
    dr.driver_rating,
    COUNT(d.delivery_id) AS total_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Ontime' THEN 1 ELSE 0 END) AS ontime_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Delayed' THEN 1 ELSE 0 END) AS delayed_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Failed' THEN 1 ELSE 0 END) AS failed_deliveries,
    ROUND(AVG(d.customer_rating_post_delivery), 2) AS avg_customer_rating,
    ROUND(AVG(d.manual_route_override_count), 2) AS avg_manual_overrides
FROM deliveries d
LEFT JOIN drivers dr
ON d.driver_id = dr.driver_id
GROUP BY dr.driver_id, dr.base_zone, dr.employment_type, dr.years_experience, dr.training_score, dr.driver_rating
ORDER BY total_deliveries DESC
")
driver_performance_view

# Vehicle performance view

vehicle_performance_view <- sqldf("
SELECT
    v.vehicle_id,
    v.vehicle_type,
    v.assigned_zone,
    v.battery_health_pct,
    v.odometer_km,
    v.maintenance_status,
    COUNT(d.delivery_id) AS total_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Ontime' THEN 1 ELSE 0 END) AS ontime_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Delayed' THEN 1 ELSE 0 END) AS delayed_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Failed' THEN 1 ELSE 0 END) AS failed_deliveries,
    ROUND(AVG(d.customer_rating_post_delivery), 2) AS avg_customer_rating,
    ROUND(AVG(d.fuel_or_charge_cost), 2) AS avg_fuel_or_charge_cost
FROM deliveries d
LEFT JOIN vehicles v
ON d.vehicle_id = v.vehicle_id
GROUP BY v.vehicle_id, v.vehicle_type, v.assigned_zone, v.battery_health_pct, v.odometer_km, v.maintenance_status
ORDER BY total_deliveries DESC
")
vehicle_performance_view

# App events performance view

app_events_performance_view <- sqldf("
SELECT
    o.service_type,
    a.event_type,
    a.device_type,
    a.zone_context,
    COUNT(a.event_id) AS total_events,
    ROUND(AVG(a.api_latency_ms), 2) AS avg_api_latency_ms,
    SUM(CASE WHEN a.success_flag = 1 THEN 1 ELSE 0 END) AS success_events,
    SUM(CASE WHEN a.success_flag = 0 THEN 1 ELSE 0 END) AS failed_events
FROM app_events a
LEFT JOIN orders o
ON a.order_id = o.order_id
GROUP BY o.service_type, a.event_type, a.device_type, a.zone_context
ORDER BY total_events DESC
")
app_events_performance_view

# Results

orders_customer_view
delivery_performance_view
incident_view
complaint_view
zone_comparison_view
hub_comparison_view
service_type_comparison_view
customer_type_performance_view
driver_performance_view
vehicle_performance_view
app_events_performance_view
```

```r
# Comparing performances

# By Zone

zone_performance <- sqldf("
SELECT
    h.zone,
    COUNT(d.delivery_id) AS total_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Ontime' THEN 1 ELSE 0 END) AS ontime_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Delayed' THEN 1 ELSE 0 END) AS delayed_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Failed' THEN 1 ELSE 0 END) AS failed_deliveries,
    ROUND(AVG(d.customer_rating_post_delivery), 2) AS avg_customer_rating,
    ROUND(AVG(d.fuel_or_charge_cost), 2) AS avg_fuel_or_charge_cost
FROM deliveries d
LEFT JOIN hubs h
ON d.hub_id = h.hub_id
GROUP BY h.zone
ORDER BY total_deliveries DESC
")
zone_performance

# By Hub

hub_performance <- sqldf("
SELECT
    h.hub_id,
    h.hub_name,
    h.zone,
    h.hub_type,
    h.capacity_score,
    COUNT(d.delivery_id) AS total_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Ontime' THEN 1 ELSE 0 END) AS ontime_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Delayed' THEN 1 ELSE 0 END) AS delayed_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Failed' THEN 1 ELSE 0 END) AS failed_deliveries,
    ROUND(AVG(d.customer_rating_post_delivery), 2) AS avg_customer_rating,
    ROUND(AVG(d.manual_route_override_count), 2) AS avg_manual_overrides
FROM deliveries d
LEFT JOIN hubs h
ON d.hub_id = h.hub_id
GROUP BY h.hub_id, h.hub_name, h.zone, h.hub_type, h.capacity_score
ORDER BY total_deliveries DESC
")
hub_performance

# By Service Type

service_type_performance <- sqldf("
SELECT
    o.service_type,
    COUNT(DISTINCT o.order_id) AS total_orders,
    ROUND(AVG(o.order_value), 2) AS avg_order_value,
    COALESCE(SUM(cbo.complaint_count), 0) AS total_complaints,
    ROUND(
        CASE
            WHEN SUM(cbo.complaint_count) > 0
            THEN SUM(cbo.total_resolution_days) / SUM(cbo.complaint_count)
        END, 2
    ) AS avg_complaint_resolution_days,
    COALESCE(SUM(dbo.delivery_count), 0) AS total_deliveries,
    ROUND(
        CASE
            WHEN SUM(dbo.delivery_count) > 0
            THEN SUM(dbo.total_customer_rating) / SUM(dbo.delivery_count)
        END, 2
    ) AS avg_customer_rating
FROM orders o
LEFT JOIN complaints_by_order cbo
ON o.order_id = cbo.order_id
LEFT JOIN deliveries_by_order dbo
ON o.order_id = dbo.order_id
GROUP BY o.service_type
ORDER BY total_orders DESC
")
service_type_performance

# By Driver

driver_performance <- sqldf("
SELECT
    dr.driver_id,
    dr.base_zone,
    dr.employment_type,
    dr.years_experience,
    dr.training_score,
    dr.driver_rating,
    COUNT(d.delivery_id) AS total_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Ontime' THEN 1 ELSE 0 END) AS ontime_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Delayed' THEN 1 ELSE 0 END) AS delayed_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Failed' THEN 1 ELSE 0 END) AS failed_deliveries,
    ROUND(AVG(d.customer_rating_post_delivery), 2) AS avg_customer_rating
FROM deliveries d
LEFT JOIN drivers dr
ON d.driver_id = dr.driver_id
GROUP BY dr.driver_id, dr.base_zone, dr.employment_type, dr.years_experience, dr.training_score, dr.driver_rating
ORDER BY total_deliveries DESC
")
driver_performance

# By Vehicle

vehicle_performance <- sqldf("
SELECT
    v.vehicle_id,
    v.vehicle_type,
    v.assigned_zone,
    v.battery_health_pct,
    v.odometer_km,
    v.maintenance_status,
    COUNT(d.delivery_id) AS total_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Ontime' THEN 1 ELSE 0 END) AS ontime_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Delayed' THEN 1 ELSE 0 END) AS delayed_deliveries,
    SUM(CASE WHEN d.delivery_status = 'Failed' THEN 1 ELSE 0 END) AS failed_deliveries,
    ROUND(AVG(d.customer_rating_post_delivery), 2) AS avg_customer_rating,
    ROUND(AVG(d.fuel_or_charge_cost), 2) AS avg_fuel_or_charge_cost
FROM deliveries d
LEFT JOIN vehicles v
ON d.vehicle_id = v.vehicle_id
GROUP BY v.vehicle_id, v.vehicle_type, v.assigned_zone, v.battery_health_pct, v.odometer_km, v.maintenance_status
ORDER BY total_deliveries DESC
")
vehicle_performance

# By Customer Type

customer_type_performance <- sqldf("
SELECT
    c.customer_type,
    COUNT(DISTINCT c.customer_id) AS total_customers,
    SUM(IFNULL(oc.order_count, 0)) AS total_orders,
    ROUND(CASE WHEN SUM(IFNULL(oc.order_count, 0)) > 0 THEN SUM(IFNULL(oc.total_order_value, 0)) / SUM(IFNULL(oc.order_count, 0)) END, 2) AS avg_order_value,
    SUM(IFNULL(cc.complaint_count, 0)) AS total_complaints,
    ROUND(CASE WHEN SUM(IFNULL(cc.complaint_count, 0)) > 0 THEN SUM(IFNULL(cc.total_resolution_days, 0)) / SUM(IFNULL(cc.complaint_count, 0)) END, 2) AS avg_complaint_resolution_days,
    ROUND(CASE WHEN SUM(IFNULL(dc.delivery_count, 0)) > 0 THEN SUM(IFNULL(dc.total_customer_rating, 0)) / SUM(IFNULL(dc.delivery_count, 0)) END, 2) AS avg_customer_rating
FROM customers c
LEFT JOIN orders_by_customer oc
ON c.customer_id = oc.customer_id
LEFT JOIN complaints_by_customer cc
ON c.customer_id = cc.customer_id
LEFT JOIN deliveries_by_customer dc
ON c.customer_id = dc.customer_id
GROUP BY c.customer_type
ORDER BY total_orders DESC
")
customer_type_performance

# By Complaint severity

complaint_severity_performance <- sqldf("
SELECT
    severity,
    COUNT(complaint_id) AS total_complaints,
    ROUND(AVG(resolution_days), 2) AS avg_resolution_days,
    ROUND(AVG(compensation_amount), 2) AS avg_compensation_amount
FROM complaints
GROUP BY severity
ORDER BY total_complaints DESC
")
complaint_severity_performance

# By Incident severity

incident_severity_performance <- sqldf("
SELECT
    severity,
    COUNT(incident_id) AS total_incidents,
    ROUND(AVG(resolved_hours), 2) AS avg_resolution_hours
FROM incidents
GROUP BY severity
ORDER BY total_incidents DESC
")
incident_severity_performance

# By App Events

app_events_performance <- sqldf("
SELECT
    zone_context,
    COUNT(event_id) AS total_events,
    ROUND(AVG(api_latency_ms), 2) AS avg_api_latency_ms,
    SUM(CASE WHEN success_flag = 1 THEN 1 ELSE 0 END) AS success_events,
    SUM(CASE WHEN success_flag = 0 THEN 1 ELSE 0 END) AS failed_events
FROM app_events
GROUP BY zone_context
ORDER BY total_events DESC
")
app_events_performance

# All results

zone_performance
hub_performance
service_type_performance
driver_performance
vehicle_performance
customer_type_performance
complaint_severity_performance
incident_severity_performance
app_events_performance
```
