```python
# Loading Functions

import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading datasets

raw_base = "https://raw.githubusercontent.com/THEX3MN/NorthStar-Dataset/main/NorthStar%20Dataset%20(Raw)"

hubs = pd.read_csv(f"{raw_base}/hubs.csv")
customers = pd.read_csv(f"{raw_base}/customers.csv")
drivers = pd.read_csv(f"{raw_base}/drivers.csv")
vehicles = pd.read_csv(f"{raw_base}/vehicles.csv")
orders = pd.read_csv(f"{raw_base}/orders.csv")
deliveries = pd.read_csv(f"{raw_base}/deliveries.csv")
incidents = pd.read_csv(f"{raw_base}/incidents.csv")
complaints = pd.read_csv(f"{raw_base}/complaints.csv")
app_events = pd.read_csv(f"{raw_base}/app_events.csv")
```

```python
# Checking for suspicious numeric values

bad_age = customers_clean[(customers_clean['age'] < 0) | (customers_clean['age'] > 100)]
print("Suspicious ages:", len(bad_age))

bad_loyalty = customers_clean[(customers_clean['loyalty_score'] < 0) | (customers_clean['loyalty_score'] > 100)]
print("Suspicious loyalty scores:", len(bad_loyalty))

bad_engagement = customers_clean[(customers_clean['app_engagement_score'] < 0) | (customers_clean['app_engagement_score'] > 100)]
print("Suspicious app engagement scores:", len(bad_engagement))

bad_experience = drivers_clean[drivers_clean['years_experience'] < 0]
print("Suspicious years of experience:", len(bad_experience))

bad_training = drivers_clean[(drivers_clean['training_score'] < 0) | (drivers_clean['training_score'] > 100)]
print("Suspicious training scores:", len(bad_training))

bad_rating = drivers_clean[(drivers_clean['driver_rating'] < 0) | (drivers_clean['driver_rating'] > 5)]
print("Suspicious driver ratings:", len(bad_rating))

bad_battery = vehicles_clean[(vehicles_clean['battery_health_pct'] < 0) | (vehicles_clean['battery_health_pct'] > 100)]
print("Suspicious battery health values:", len(bad_battery))

bad_window = orders_clean[orders_clean['promised_window_hours'] <= 0]
print("Suspicious promised window hours:", len(bad_window))

bad_order_value = orders_clean[orders_clean['order_value'] < 0]
print("Suspicious order values:", len(bad_order_value))

bad_distance = deliveries_clean[deliveries_clean['route_distance_km'] <= 0]
print("Suspicious route distances:", len(bad_distance))

bad_override = deliveries_clean[deliveries_clean['manual_route_override_count'] < 0]
print("Suspicious manual route override counts:", len(bad_override))

bad_completion_flag = deliveries_clean[~deliveries_clean['proof_of_completion_missing'].isin([0, 1])]
print("Suspicious proof_of_completion_missing values:", len(bad_completion_flag))

bad_active_flag = drivers_clean[~drivers_clean['active_flag'].isin([0, 1])]
print("Suspicious active_flag values:", len(bad_active_flag))

bad_special_flag = orders_clean[~orders_clean['special_handling_flag'].isin([0, 1])]
print("Suspicious special_handling_flag values:", len(bad_special_flag))

bad_success_flag = app_events_clean[~app_events_clean['success_flag'].isin([0, 1])]
print("Suspicious success_flag values:", len(bad_success_flag))

bad_resolved_hours = incidents_clean[incidents_clean['resolved_hours'] < 0]
print("Suspicious resolved hours:", len(bad_resolved_hours))

bad_resolution_days = complaints_clean[complaints_clean['resolution_days'] < 0]
print("Suspicious resolution days:", len(bad_resolution_days))

bad_compensation = complaints_clean[complaints_clean['compensation_amount'] < 0]
print("Suspicious compensation amounts:", len(bad_compensation))

# Checking for suspicious date order in deliveries

bad_delivery_times = deliveries_clean[
    deliveries_clean['delivery_completed_at'].notna() &
    (deliveries_clean['delivery_completed_at'] < deliveries_clean['dispatch_time'])
]
print("Deliveries completed before dispatch:", len(bad_delivery_times))
```

```python
# Identify and handle invalid delivery timestamps

deliveries_clean['invalid_time_flag'] = (
    deliveries_clean['delivery_completed_at'].notna() &
    (deliveries_clean['delivery_completed_at'] < deliveries_clean['dispatch_time'])
)
print(deliveries_clean['invalid_time_flag'].sum())
deliveries_valid_time = deliveries_clean[deliveries_clean['invalid_time_flag'] == False]
```

```python
 # Missing value handling

customers_clean['loyalty_score'] = customers_clean['loyalty_score'].fillna(customers_clean['loyalty_score'].median())
customers_clean['preferred_channel'] = customers_clean['preferred_channel'].fillna('Unknown')

drivers_clean['training_score'] = drivers_clean['training_score'].fillna(drivers_clean['training_score'].median())

vehicles_clean['battery_health_pct'] = vehicles_clean['battery_health_pct'].fillna(vehicles_clean['battery_health_pct'].median())

orders_clean['booking_channel'] = orders_clean['booking_channel'].fillna('Unknown')

complaints_clean['compensation_amount'] = complaints_clean['compensation_amount'].fillna(0)

orders_clean['high_priority_flag'] = orders_clean['priority_level'].isin(['High', 'Critical']).astype(int)
```

```python
# Creating analysis features

# Ensuring filtered deliveries dataframe is an independent copy
deliveries_valid_time = deliveries_valid_time.copy()

# High cost deliveries
deliveries_valid_time.loc[:, 'high_cost_flag'] = np.where(
    deliveries_valid_time['fuel_or_charge_cost'] > deliveries_valid_time['fuel_or_charge_cost'].median(),
    1,
    0
)

# High delay risk deliveries
deliveries_valid_time.loc[:, 'high_delay_risk_flag'] = np.where(
    (deliveries_valid_time['manual_route_override_count'] > 0) |
    (deliveries_valid_time['delivery_status'] == 'Delayed'),
    1,
    0
)

# Customers with above average loyalty
customers_clean.loc[:, 'above_avg_loyalty_flag'] = np.where(
    customers_clean['loyalty_score'] > customers_clean['loyalty_score'].mean(),
    1,
    0
)

# Vehicles with low battery health
vehicles_clean.loc[:, 'low_battery_flag'] = np.where(
    vehicles_clean['battery_health_pct'] < vehicles_clean['battery_health_pct'].median(),
    1,
    0
)

# Customer features
customers_clean.loc[:, 'high_loyalty_flag'] = (customers_clean['loyalty_score'] >= customers_clean['loyalty_score'].median()).astype(int)
customers_clean.loc[:, 'high_engagement_flag'] = (customers_clean['app_engagement_score'] >= customers_clean['app_engagement_score'].median()).astype(int)

# Driver features
drivers_clean.loc[:, 'experienced_driver_flag'] = (drivers_clean['years_experience'] >= drivers_clean['years_experience'].median()).astype(int)
drivers_clean.loc[:, 'high_training_flag'] = (drivers_clean['training_score'] >= drivers_clean['training_score'].median()).astype(int)

# Vehicle features
vehicles_clean.loc[:, 'good_battery_flag'] = (vehicles_clean['battery_health_pct'] >= vehicles_clean['battery_health_pct'].median()).astype(int)
vehicles_clean.loc[:, 'high_mileage_flag'] = (vehicles_clean['odometer_km'] >= vehicles_clean['odometer_km'].median()).astype(int)

# Order features
orders_clean.loc[:, 'high_value_flag'] = (orders_clean['order_value'] >= orders_clean['order_value'].median()).astype(int)
orders_clean.loc[:, 'priority_flag'] = (orders_clean['priority_level'] == 'High').astype(int)
orders_clean.loc[:, 'special_handling_flag'] = orders_clean['special_handling_flag'].astype(int)

# Delivery features
deliveries_valid_time.loc[:, 'delivery_duration_hours'] = (
    deliveries_valid_time['delivery_completed_at'] - deliveries_valid_time['dispatch_time']
).dt.total_seconds() / 3600

deliveries_valid_time.loc[:, 'delay_flag'] = (deliveries_valid_time['delivery_status'] == 'Delayed').astype(int)
deliveries_valid_time.loc[:, 'failure_flag'] = (deliveries_valid_time['delivery_status'] == 'Failed').astype(int)
deliveries_valid_time.loc[:, 'ontime_flag'] = (deliveries_valid_time['delivery_status'] == 'Ontime').astype(int)
deliveries_valid_time.loc[:, 'override_flag'] = (deliveries_valid_time['manual_route_override_count'] > 0).astype(int)
deliveries_valid_time.loc[:, 'low_rating_flag'] = (deliveries_valid_time['customer_rating_post_delivery'] <= 3).astype(int)

# Avoiding division by zero for route distance
deliveries_valid_time.loc[:, 'cost_per_km'] = np.nan
deliveries_valid_time.loc[deliveries_valid_time['route_distance_km'] > 0, 'cost_per_km'] = (
    deliveries_valid_time.loc[deliveries_valid_time['route_distance_km'] > 0, 'fuel_or_charge_cost'] /
    deliveries_valid_time.loc[deliveries_valid_time['route_distance_km'] > 0, 'route_distance_km']
)

# Incident features
incidents_clean.loc[:, 'high_severity_flag'] = (incidents_clean['severity'] == 'High').astype(int)
incidents_clean.loc[:, 'slow_resolution_flag'] = (incidents_clean['resolved_hours'] > incidents_clean['resolved_hours'].median()).astype(int)

# Complaint features
complaints_clean.loc[:, 'open_complaint_flag'] = (complaints_clean['status'] != 'Resolved').astype(int)
complaints_clean.loc[:, 'high_compensation_flag'] = (complaints_clean['compensation_amount'] >= complaints_clean['compensation_amount'].median()).astype(int)
complaints_clean.loc[:, 'long_resolution_flag'] = (complaints_clean['resolution_days'] > complaints_clean['resolution_days'].median()).astype(int)

# App event features
app_events_clean.loc[:, 'high_latency_flag'] = (app_events_clean['api_latency_ms'] >= app_events_clean['api_latency_ms'].median()).astype(int)
app_events_clean.loc[:, 'failed_event_flag'] = (app_events_clean['success_flag'] == 0).astype(int)
app_events_clean.loc[:, 'event_date'] = app_events_clean['event_timestamp'].dt.date
```
